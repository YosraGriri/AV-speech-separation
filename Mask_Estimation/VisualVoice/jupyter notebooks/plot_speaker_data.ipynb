```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simulated data for utterance lengths
utterance_lengths = np.random.normal(12, 2, size=1000)
bins = np.arange(4, 21, 2)  # Bins for the histogram

# Create histogram
plt.figure(figsize=(6, 4))
plt.hist(utterance_lengths, bins=bins, color='skyblue', edgecolor='black')
plt.title('Utterance Lengths')
plt.xlabel('Length (sec)')
plt.ylabel('Frequency')
plt.savefig('plots/ydgfxd.pdf', format="pdf", dpi=300, bbox_inches="tight")

plt.show()
plt.close()
```

```python
# Data for the gender distribution pie chart
gender_labels = ['Male', 'Female']
gender_sizes = [61, 39]  # Percentages
gender_colors = ['#80BCBD', '#AAD9BB']  # Using HEX colors

# Plot gender distribution pie chart
plt.figure(figsize=(6, 4))
plt.pie(gender_sizes, labels=gender_labels, colors=gender_colors, autopct='%1.1f%%', startangle=140)
plt.savefig('plots/Gender_Distribution.pdf', format="pdf", dpi=300, bbox_inches="tight")
plt.title('Gender Distribution')
plt.show()
plt.close()
```

```python
# Data for the nationality distribution pie chart
nationality_labels = ['USA', 'UK', 'Unknown', 'Germany', 'India', 'France', 'Other']
nationality_sizes = [29, 10, 6, 6, 6, 6, 37]  # Percentages
nationality_colors = ['#386641', '#6a994e', '#d8f3dc', '#80BCBD', '#40916c', '#AAD9BB', '#34a0a4']  # Using HEX colors

# Plot nationality distribution pie chart
plt.figure(figsize=(6, 4))
plt.pie(nationality_sizes, labels=nationality_labels, colors=nationality_colors, autopct='%1.1f%%', startangle=140)

plt.savefig('plots/Nationality_Distribution.pdf', format="pdf", dpi=300, bbox_inches="tight")
plt.title('Nationality Distribution')
plt.show()
plt.close()
```
