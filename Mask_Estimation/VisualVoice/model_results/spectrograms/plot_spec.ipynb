```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display  # This is helpful for plotting spectrograms with librosa
```

```python
# Load the magnitude spectrogram from .npy file
magnitude_spectrogram_1= np.load('pred_spec_1_real.npy')
magnitude_spectrogram_2= np.load('pred_spec_2_real.npy')
magnitude_spec_mix = np.load('magnitude_spec_mix.npy')
```

```python
# Convert the magnitude spectrogram to dB
db_spectrogram = librosa.amplitude_to_db(magnitude_spectrogram_1, ref=np.max)

# Normalize the dB spectrogram to be between -1 and 1
db_spectrogram_normalized = 2 * ((db_spectrogram - db_spectrogram.min()) / (db_spectrogram.max() - db_spectrogram.min())) - 1

# Plot the normalized dB spectrogram
plt.figure(figsize=(10, 5))
librosa.display.specshow(db_spectrogram, sr=16000, hop_length=160, x_axis='time', y_axis='linear', cmap='magma')
plt.colorbar(format='%+2.0f dB')
plt.title('Normalized Magnitude Spectrogram')
plt.tight_layout()
plt.show()
```
